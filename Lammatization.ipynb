```python
#O&A, Chatbots, TextSummarizer etc.
from nltk.stem import WordNetLemmatizer
```

```python
lemmatizer = WordNetLemmatizer()
```

```python
'''
POS- Noun - n
Verb - v
adjective - a
adverb - r
'''

lemmatizer.lemmatize("going", pos="v")
```

```python
words = ["eating", "eats", "eat", "eaten", "writes", "writing", "write", "programming", "program", "programs", "history"]
```

```python
for word in words:
    print(word+"--->"+lemmatizer.lemmatize(word, pos="v"))
```
