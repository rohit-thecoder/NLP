About Dataset

Context This is a small subset of dataset of Book reviews from Amazon Kindle Store category.

Content 5-core dataset of product reviews from Amazon Kindle Store category from May 1996 - July 2014. Contains total of 982619 entries. Each reviewer has at least 5 reviews and each product has at least 5 reviews in this dataset. Columns

asin - ID of the product, like B000FA64PK
helpful - helpfulness rating of the review - example: 2/3.
overall - rating of the product.
reviewText - text of the review (heading).
reviewTime - time of the review (raw).
reviewerID - ID of the reviewer, like A3SPTOKDG7WBLN
reviewerName - name of the reviewer.
summary - summary of the review (description).
unixReviewTime - unix timestamp.
Acknowledgements This dataset is taken from Amazon product data, Julian McAuley, UCSD website. http://jmcauley.ucsd.edu/data/amazon/

License to the data files belong to them.

```python
#Load Dataset
import pandas as pd
data = pd.read_csv("Kindle Review/all_kindle_review.csv")
data.head()
```

```python
df = data[["reviewText", "rating"]]
df.head()
```

```python
df.shape
```

```python
##Missing values
df.isnull().sum()
```

```python
#Let's see how many unique ratings are there
df["rating"].unique()
```

```python
df["rating"].value_counts()
```

```python
#Preprocessing and Cleaning

#Positive review as 1 and negative review as 0
df["rating"] = df["rating"].apply(lambda x:0 if x<3 else 1)
```

```python
df.head()
```

```python
df["rating"].unique()
```

```python
df["rating"].value_counts()
```

```python
#1. Lower all the case
df["reviewText"] = df["reviewText"].str.lower()
```

```python
df.head()
```

```python
import re
import nltk
from nltk.corpus import stopwords
nltk.download("stopwords")
```

```python
from bs4 import BeautifulSoup
```

```python
## Removing special characters
df['reviewText']=df['reviewText'].apply(lambda x:re.sub('[^a-z A-z 0-9-]+', '',x))
## Remove the stopswords
df['reviewText']=df['reviewText'].apply(lambda x:" ".join([y for y in x.split() if y not in stopwords.words('english')]))
## Remove url 
df['reviewText']=df['reviewText'].apply(lambda x: re.sub(r'(http|https|ftp|ssh)://([\w_-]+(?:(?:\.[\w_-]+)+))([\w.,@?^=%&:/~+#-]*[\w@?^=%&/~+#-])?', '' , str(x)))
## Remove html tags
df['reviewText']=df['reviewText'].apply(lambda x: BeautifulSoup(x, 'lxml').get_text())
## Remove any additional spaces
df['reviewText']=df['reviewText'].apply(lambda x: " ".join(x.split()))
```

```python
df.head()
```

```python
##Applying Lemmatizer
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
```

```python
def lemmatize_word(text):
    return " ".join[(lemmatizer.lemmatize(word) for word in text.split())]
```

```python
df["reviewText"] = df["reviewText"].apply(lambda x:lemmatize_word(x))
```

```python
df.head()
```

```python
#Train Test Split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df["reviewText"], df["ratings"], test_size=0.20)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
bow = CountVectorizer()
X_train_bow = bow.fit_transform(X_train).toarray()
X_test_bow = bow.transform(X_test).toarray()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer()
X_train_tfidf = tfidf.fit_transform(X_train).toarray()
X_test_tfidf = tfidf.transform(X_test).toarray()
```

```python
X_train_bow
```

```python
X_train_tfidf
```

```python
from sklearn.naive_bayes import GaussianNB
nb_model_bow = GaussianNB().fit(X_train_bow, y_train)
tfidf_model_bow = GaussianNB().fit(X_train_tfidf, y_train)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
```

```python
y_pred_bow = nb_model_bow.predict(X_test_bow)
y_pred_tfidf = nb_model_bow.predict(X_test_tfidf)
```

```python
confusion_matrix(y_test, y_pred_bow)
```

```python
print("BOW accuracy",accuracy_score(y_test, y_pred_bow))
```

```python
confusion_matrix(y_test, y_pred_tfidf)
```

```python
print("TF-IDF accuracy",accuracy_score(y_test, y_pred_tfidf))
```
